# 07 Clean

Очищення згенерованих файлів з `07a`, `07b`, `07c`, `07d`.

Логіка:
- спочатку показуємо, які файли/папки знайдені
- видаляємо тільки якщо `CONFIRM_DELETE = True`
- базові вхідні файли на кшталт `accessibility_scores.csv` не чіпаємо


```python
from pathlib import Path
import shutil

ROOT = Path('..').resolve()

# Спочатку лишаємо dry-run, щоб випадково нічого не стерти.
CONFIRM_DELETE = True

# Артефакти, які створюють 07a / 07b / 07c / 07d.
TARGETS = [
    # ROOT / 'data/processed/buildings.parquet',
    ROOT / 'data/processed/stop_to_bld_short.parquet',
    ROOT / 'data/processed/stop_to_bld_long.parquet',
    ROOT / 'data/processed/stop_to_fac_exit.parquet',
    ROOT / 'data/processed/stop_reachability_peak.parquet',
    ROOT / 'data/processed/stop_reachability_offpeak.parquet',
    ROOT / 'data/processed/stop_reachability_peak_reversed.parquet',
    ROOT / 'data/processed/stop_reachability_offpeak_reversed.parquet',
    ROOT / 'data/processed/wait_times_peak.parquet',
    ROOT / 'data/processed/wait_times_offpeak.parquet',
    ROOT / 'data/processed/catchment_results.csv',
    ROOT / 'data/processed/catchment_buildings.parquet',
    ROOT / 'data/processed/map_data.json',
    ROOT / 'data/processed/map_buildings',
    ROOT / 'outputs/15_catchment_groups.png',
    ROOT / 'outputs/map_temp.html',
    ROOT / 'outputs/map_catchment_interactive.html',
]

existing = [p for p in TARGETS if p.exists()]
missing = [p for p in TARGETS if not p.exists()]

print('Знайдено для очищення:')
for path in existing:
    rel = path.relative_to(ROOT)
    kind = 'DIR ' if path.is_dir() else 'FILE'
    print(f'  [{kind}] {rel}')

print(f'\nВідсутні: {len(missing)}')
print(f'Знайдено: {len(existing)}')

if not CONFIRM_DELETE:
    print('\nDry-run режим. Щоб реально видалити файли, постав CONFIRM_DELETE = True і запусти клітинку ще раз.')
else:
    deleted = []
    errors = []

    for path in existing:
        try:
            if path.is_dir():
                shutil.rmtree(path)
            else:
                path.unlink()
            deleted.append(path)
        except Exception as exc:
            errors.append((path, exc))

    print('\nВидалено:')
    for path in deleted:
        print(f'  - {path.relative_to(ROOT)}')

    if errors:
        print('\nПомилки:')
        for path, exc in errors:
            print(f'  - {path.relative_to(ROOT)} -> {exc}')
    else:
        print('\nОчищення завершено без помилок.')
```
